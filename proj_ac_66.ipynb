```python
import pandas as pd
```

```python
df_awards_players = pd.read_csv('data/awards_players.csv')
df_coaches = pd.read_csv('data/coaches.csv')
df_players_teams = pd.read_csv('data/players_teams.csv')
df_players = pd.read_csv('data/players.csv')
df_series_post = pd.read_csv('data/series_post.csv')
df_teams = pd.read_csv('data/teams.csv')
df_teams_post = pd.read_csv('data/teams_post.csv')
```

```python
df_teams.head()
```

```python

# Step 3: Identify columns with only NaN values or all the same values
columns_to_drop = []

# Check for columns with all NaN values
columns_to_drop = []
for col in df_teams.columns:
    unique_values = df_teams[col].unique()
    if len(unique_values) == 1 or all(pd.isna(x) for x in unique_values):
        columns_to_drop.append(col)

# Step 2: Drop identified columns
df_teams.drop(columns=columns_to_drop, inplace=True)

# Step 4: Delete the identified columns
print(columns_to_drop)

```

```python
df_teams.isnull().sum()
```

```python
print("Total number of duplicate rows: ", df_teams.duplicated().sum())
df_teams.drop_duplicates(inplace=True)
print("Total number of duplicate rows: ", df_teams.duplicated().sum())
```

```python
df_teams = pd.merge(df_teams, df_coaches, on=['year', 'tmID'], how='left')
df_teams = pd.merge(df_teams, df_players_teams, on=['year',  'tmID'], how='left')
df_teams = pd.merge(df_teams, df_series_post, left_on=['year', 'tmID'], right_on=['year', 'tmIDWinner'], how='left')
df_teams = pd.merge(df_teams, df_teams_post, on=['year', 'tmID'], how='left')
```
